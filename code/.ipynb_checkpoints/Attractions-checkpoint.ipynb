```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from functools import reduce
```

```python
folderpath = r"../hourly_data/" 
rides = []             # List of names to use to name the columns in the dataframe
attraction = []        # List of dataframes of imported csv files
data = {}
for file in os.listdir(folderpath):
    if file.endswith('.csv'):
        ride =  file.replace('.csv', '') 
        data[ride] = pd.read_csv(os.path.join(folderpath, file), dtype = object) 
        rides.append(ride)
        attraction.append((data[ride]))
```

```python
attractions = reduce(lambda  left,right: pd.merge(left,right,on=['datetime'],
                                            how='outer'), attraction).fillna('na')
attractions.columns = attractions.columns[:1].tolist() + rides
attractions.sort_index()
```
