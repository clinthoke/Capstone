```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
slinky_dog = pd.read_csv("../cleaned_data/slinky_dog.csv")
slinky_dog.info()
```

```python
slinky_dog
```

```python
pd.crosstab(slinky_dog.date, slinky_dog.SPOSTMIN)
```

```python
slinky_dog['date'] = pd.to_datetime(slinky_dog["date"])
slinky_dog.dtypes
```

```python
slinky_dog
```

```python
# pd.crosstab(slinky_dog.date, slinky_dog.SPOSTMIN)
```

```python
slinky_dog['datetime'] = pd.to_datetime(slinky_dog["datetime"])
slinky_dog.dtypes
```

```python
# set date column to be index
slinky_dog.set_index('datetime', inplace = True)
slinky_dog.head()
```

```python
slinky_dog.resample("H").mean().info()
```

```python
slinky_dog.resample("H").mean()
```

```python
slinky_dog = slinky_dog.resample("H").mean().dropna()
slinky_dog
```

```python
slinky_dog.reset_index(level=0, inplace=True)
```

```python
slinky_dog
```

```python
slinky_dog.to_csv('../hourly_data/slinky_dog.csv', index = False)
```
