```python
import pandas as pd
import os
```

```python
basepath = "../cleaned_data/"
futurepath = "../hourly_data/"
for filename in os.listdir(basepath):
    if filename.endswith(".csv"): 
        ride = os.path.join("." +filename)
        ride = pd.read_csv(basepath + filename)
        print(ride.dtypes)
        ride['date'] = pd.to_datetime(ride["date"])
        ride['datetime'] = pd.to_datetime(ride["datetime"])
        ride = ride[~(ride['date'] < '2015-01-01')]
        ride = ride[~(ride['date'] > '2019-12-31')]
        print(ride.dtypes)
        ride.set_index('datetime', inplace = True)
        ride = ride.resample("H").mean().round().dropna()
        ride.reset_index(level=0, inplace=True)
        print(ride.info())        
        ride.to_csv('../hourly_data/' + filename, index = False)
        print(f'{filename} is averaged by hour and is now in the hourly_data folder')        
```
